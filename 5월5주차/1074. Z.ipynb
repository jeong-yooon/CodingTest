```python
#   
def quad_tree(N, x, y):
    global arr
    count = 0
    
    #  
    for i in range(x, x + N):
        for j in range(y, y + N):
            quad_tree(N // 2, x, y) #  
            quad_tree(N // 2, x, y + N // 2) #  
            quad_tree(N // 2, x + N // 2, y) #  
            quad_tree(N // 2, x + N // 2, y + N // 2) #  
            arr[x][y] = count
            count += 1
            return


#  
N, r, c = map(int, input().split())
arr = []
for i in range(2**N):
    arr.append([0]*(2**N)

quad_tree((2**N), 0, 0)
    
print(arr[r][c])
```

```python
#  
N, r, c = map(int, input().split())

answer = 0

while N != 0: #  1  Z  

    N -= 1 #    

    #  
    if r < 2 ** N and c < 2 ** N:
        answer += ( 2 ** N ) * ( 2 ** N ) * 0

    #  
    elif r < 2 ** N and c >= 2 ** N: 
        answer += ( 2 ** N ) * ( 2 ** N ) * 1
        c -= ( 2 ** N )
        
    #    
    elif r >= 2 ** N and c < 2 ** N: 
        answer += ( 2 ** N ) * ( 2 ** N ) * 2
        r -= ( 2 ** N )
        
    #    
    else:
        answer += ( 2 ** N ) * ( 2 ** N ) * 3
        r -= ( 2 ** N )
        c -= ( 2 ** N )
    
print(answer)
```
