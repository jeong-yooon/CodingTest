## 문제
코로나 바이러스로 인해 H 대학은 비대면 강의를 실시하고 있다. 조교를 담당하게 된 지환이는 출석체크 방식을 바꾸려고 한다.

학생들은 접속 순서대로 3번부터 N + 2번까지 입장 번호를 받게 된다.

지환이가 한 학생에게 출석 코드를 보내게 되면, 해당 학생은 본인의 입장 번호의 배수인 학생들에게 출석 코드를 보내어 해당 강의의 출석을 할 수 있게끔 한다.

하지만, K명의 졸고 있는 학생들은 출석 코드를 제출하지 않고, 다른 학생들에게 보내지 않는다.

지환이는 무작위로 한 명의 학생에게 출석 코드를 보내는 행위를 Q번 반복한 뒤, 출석부 정리를 위해 특정 구간의 입장 번호를 받은 학생들 중에서 출석이 되지 않은 학생들의 수를 구하고 싶다.

많은 인원을 담당해서 바쁜 지환이를 위해 프로그램을 만들어주자!

### 입력
1번째 줄에 학생의 수 N, 졸고 있는 학생의 수 K, 지환이가 출석 코드를 보낼 학생의 수 Q, 주어질 구간의 수 M이 주어진다. (1 ≤ K, Q ≤ N ≤ 5,000, 1 ≤ M ≤ 50,000)

2번째 줄과 3번째 줄에 각각 K명의 졸고 있는 학생의 입장 번호들과 Q명의 출석 코드를 받을 학생의 입장 번호들이 주어진다.

4번째 줄부터 M개의 줄 동안 구간의 범위 S, E가 공백을 사이에 두고 주어진다. (3 ≤ S < E ≤ N + 2)

### 출력
M개의 줄에 걸쳐서 각 구간에 대해서 출석이 되지 않은 학생들의 수를 출력하라.

### 예제 입력 1 
10 1 3 1\
7\
3 5 7\
3 12
### 예제 출력 1 
4

입장 번호 3번부터 12번까지의 구간에서 입장 번호 4, 8, 11번이 출석 코드를 받지 못했고, 7번은 출석 코드를 받았으나 조느라 출석하지 못했다.

### 예제 입력 2 
50 4 5 1\
24 15 27 43\
3 4 6 20 25\
3 52
### 예제 출력 2 
25

#### 알고리즘 분류
- 누적 합
#### 풀이
- 학생번호 3 ~ N+2
- 각 학생은 자신의 배수번호인 학생들에게 출석 코드를 보낸다.
- K명은 졸아서 제출X, 다른 학생에게 전달 X
- Q번 반복

[예제 입력1]
- 7번 학생이 졸고있으니 세번째 줄의 3 5 7 중에 3 5 만 사용하여 진행하여 출석 진행
- 출석 학생 번호: 3 5 6 9 10 12
- 미 출석 학생 번호: 4 7 8 11

[예제 입력2]
- 24 15 27 43 번호의 학생들이 졸고 있으니 배수값인 28 30 45 번호의 학생들도 출석 X
- 출석 학생 번호: 3 4 6 8 9 12 16 18 20 21 25 32 33 36 39 40 42 44 48 50 51 52
- 미출석 학생 번호: 5 7 10 11 13 14 15 17 19 22 23 24 26 27 28 29 30 31 34 35 37 38 41 43 45 46 47 49

- 먼저 K의 배수가 되는 학생들을 모두 정답 배열에 담는다.
- M 범위에서 Q의 배수가 되는 배열을 제외한 수들을 정답 배열에 담는다.
- 중복값

```python
N, K, Q, M = map(int, input().split())

# set은 중복된 값을 허용하지 않고 순서가 없는 항목들의 모음을 생성
sleep = set(map(int, input().split())) # 졸고 있는 학생
students = set(map(int, input().split())) # 출석 코드 받을 학생
attend = [False for _ in range(N + 3)] # 체크 리스트 선언

# 1 출석 체크
for s in (students - sleep): # 출석 코드 받을 학생 중 조는 학생을 제외하고 탐색
    for i in range(s, N + 3, s): # 배수씩 탐색하여라
        if i not in sleep: # 탐색 인덱스가 졸고 있는 학생이 아니면
            attend[i] = True # 리스트에 적용
            
# 2 미출석자 누적합 구하기
memo = [0 for _ in range(N + 3)] # 0으로 채워진 리스트 선언
for i in range(3, N + 3): # 전체 학생 수만큼 순회하며
    if not attend[i]:
        memo[i] = memo[i - 1] + 1 # 출석이 안된 번호라면 이전 인덱스값에 1을 더해 저장
    else:
        memo[i] = memo[i - 1] # 출석된 번호라면 이전 인덱스값 그대로 저장

# 3 출력
for _ in range(M):
    s, e = map(int, input().split()) # 범위 입력 받기
    ans = memo[e] - memo[s - 1] # 범위 만큼만 누적값 가져오기
print(ans)
```
